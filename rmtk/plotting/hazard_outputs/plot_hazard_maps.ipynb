#Hazard Maps


This IPython notebook allows users to plot hazard maps generated from probabilistic seismic hazard analysis.

Please specify the path of the xml file containing the hazard map results. The bounding box defining the extents of the map is estimated directly from the list of sites in the results file. The figure generated can be saved to an image file by providing a filename to the `output_file` parameter below. The formats png, pdf, ps, eps and svg are currently supported.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from plot_hazard_outputs import HazardMap
```

```python
hazard_map_file = "../sample_outputs/hazard/hazard_map.xml"
hazard_map = HazardMap(hazard_map_file)
```
